```python
import numpy as np

x = np.array([1,2,3,4,5,6,7,8])
y=x

plt.figure()
plt.scatter(x,y)
```

```python
x = np.array([1,2,3,4,5,6,7,8])
y=x

plt.figure()
plt.scatter(x,y)

colors = (len(x)-1)*['green']
colors.append('red')

plt.figure()
plt.scatter(x,y,s=100,c=colors)
```

```python
plt.xlabel('x')
plt.ylabel('y')
plt.title('Relationship')
```
