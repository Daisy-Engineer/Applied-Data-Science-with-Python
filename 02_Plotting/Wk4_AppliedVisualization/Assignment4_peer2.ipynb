```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_excel("Roman Apostolic Catholic, Brazil's Regions, 2000 and 2010.xlsx")
```

```python
df1
```

```python
df1['% Catholic 2000'] = (df1['Roman Apostolic Catholic 2000']/df1['Total 2000'].sum())*100
df1['% Catholic 2000'] = df1['% Catholic 2000'].apply(lambda x: round(float(x),2))

df1['% Catholic 2010'] = (df1['Roman Apostolic Catholic 2010']/df1['Total 2010'].sum())*100
df1['% Catholic 2010'] = df1['% Catholic 2010'].apply(lambda x: round(float(x),2))
```

```python
df1
```

```python
df2 = pd.read_excel("Evangelicals, Brazil's Regions, 2000 and 2010.xlsx")
```

```python
df2
```

```python
df2['% Evangelicals 2000'] = (df2['Evangelicals 2000']/df2['Total 2000'].sum())*100
df2['% Evangelicals 2000'] = df2['% Evangelicals 2000'].apply(lambda x: round(float(x),2))

df2['% Evangelicals 2010'] = (df2['Evangelicals 2010']/df2['Total 2010'].sum())*100
df2['% Evangelicals 2010'] = df2['% Evangelicals 2010'].apply(lambda x: round(float(x),2))
```

```python
df2
```

```python
fig = plt.figure()

xvals = range(df1.shape[0])
regions = ['          North', '          Northeast', '          Southeast', '          South', '          Midwest']

plt.bar(xvals, df1['% Catholic 2000'], width = 0.3, color='lightcoral', alpha=0.7, label='Catholics in 2000')
plt.bar(xvals, df2['% Evangelicals 2000'], width = 0.3, bottom=df1['% Catholic 2000'], color='brown', alpha=0.7, label='Evangelicals in 2000')

new_xvals = []
for item in xvals:
    new_xvals.append(item+0.3)
plt.bar(new_xvals, df1['% Catholic 2010'], width = 0.3 ,color='yellowgreen', alpha=0.7, label='Catholics in 2010')
plt.bar(new_xvals, df2['% Evangelicals 2010'], width = 0.3, bottom=df1['% Catholic 2010'], color='darkolivegreen', alpha=0.7, label='Evangelicals in 2010')

plt.xticks(xvals, regions)

ax = plt.gca()
for tick in ax.axes.get_xticklines():
    tick.set_visible(False)

plt.ylim(0, 40)

plt.ylabel('% of Total Population', fontsize=12)
plt.title("Percentage of Catholics and Evangelicals \nin Brazil's Regions in 2000 and 2010 Census\n\n", fontsize=16)

plt.subplots_adjust(top=1.2)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.legend(loc=1, title='Legend')
plt.subplots_adjust(right=1.5)
```
